**`Mielage Predicition -Regression Analysis`**

**`Objective`**

The objective of mileage prediction using regression analysis is to estimate a vehicle's fuel efficiency (mileage) based on factors like engine size, weight, or horsepower, and understand how these factors affect fuel consumption.
  

**`Data Source `**
This data set is was taken from the statLib Library which is maintained at Carneign mellon University .The dataset was used in 1983 Americsn Statistical association exposition

Remove Missing Value

```python
df=df.dropna()
```

**Data Visualization**

```python
sns.pairplot(df,x_vars=['displacement','horsepower','weight','acceleration','mpg'],y_vars=['mpg']);
```

**Define Target variable y and features X **

```python
df.columns
```

```python
y=df['mpg']
```

```python
y.shape
```

```python
X=df[['displacement', 'horsepower', 'weight', 'acceleration']]
```

```python
X.shape
```

```python
X
```

**Scaling data**

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss=StandardScaler()
```

```python
X=ss.fit_transform(X)
```

```python
X
```

```python
pd.DataFrame(X).describe()
```

**Train Test Split Data **

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test= train_test_split(X,y,train_size=0.7,random_state=2529)
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

**Linear Regression Model **

```python
from sklearn.linear_model import LinearRegression
```

```python

lr=LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
lr.intercept_
```

```python
lr.coef_
```

**Mileage =23.4-1.050Displacment-1.68Horespower-4.10Weight-0.115Acclearation+error**

**Predict Test Data **

```python
y_pred=lr.predict(X_test)
```

```python
y_pred
```

**Model Accuracy **

```python
from sklearn.metrics import mean_absolute_error,mean_absolute_percentage_error,r2_score
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
r2_score(y_test,y_pred)
```

**Polynominal Regression**

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly=PolynomialFeatures(degree=2,interaction_only =True,include_bias=False)
```

```python
X_train2=poly.fit_transform(X_train)
```

```python
X_test2=poly.fit_transform(X_test)
```

```python
lr.fit(X_train2,y_train)
```

```python
lr.intercept_
```

```python
lr.coef_
```

```python
y_pred_poly=lr.predict(X_test2)
```

**Model Accuracy**

```python
from sklearn.metrics import mean_absolute_error,mean_absolute_percentage_error,r2_score
```

```python
mean_absolute_error(y_test,y_pred_poly)
```

```python
r2_score(y_test,y_pred_poly)
```
